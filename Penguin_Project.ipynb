```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data=pd.read_csv('penguins.csv')

plt.figure(figsize=(16,10))
Species=data.species.value_counts()
Species.plot(kind='pie', autopct="%.2f%%")
```

```python
plt.figure(figsize=(16,10))
Sex=data.sex.value_counts()
Sex.plot(kind='pie',autopct="%.2f%%")
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x='culmen_length_mm', y='culmen_depth_mm', data=data, hue="species")
plt.title('Data Culmen Length vs Culmen Depth', color='red', fontsize=20)
plt.xlabel('Culmen Length (mm)', fontsize=15)
plt.ylabel('Culmen Depth (mm)', fontsize=15)
plt.legend(loc="lower right", fontsize=15)
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(x='species', y='culmen_length_mm', data=data, hue="island")
plt.title('Data Species vs Culmen length', color='red', fontsize=20)
plt.xlabel('Species', fontsize=15)
plt.ylabel('Culmen Length (mm)', fontsize=15)
plt.legend(loc="lower right", fontsize=15)
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(x='species', y='culmen_depth_mm', data=data, hue="island")
plt.title('Data Species vs Culmen Depth', color='red', fontsize=20)
plt.xlabel('Species', fontsize=15)
plt.ylabel('Culmen Depth (mm)', fontsize=15)
plt.legend(loc="lower right", fontsize=15)
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=data, hue="species")
plt.title('Data Flipper Length vs Body Mass',color='red', fontsize=20)
plt.xlabel('Flipper Length (mm)', fontsize=15)
plt.ylabel('Body Mass (g)', fontsize=15)
plt.legend(loc="lower right", fontsize=15)
```
